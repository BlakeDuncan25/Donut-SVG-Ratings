```python
# library
import matplotlib.pyplot as plt
```

```python
rating = 1

while rating < 101:
    # Create a circle for the center of the plot
    size=[rating,(100-rating)]
    if rating >= 85:
        colors=['red','white']
    elif rating >= 65:
        colors=['orange','white']
    elif rating >=40:
        colors=['yellow','white']
    else:
        colors=['green','white']
    plt.pie(size, colors=colors, startangle=90, counterclock=False)
    my_circle=plt.Circle( (0,0), 0.9, color='white')

    # Give color names
    p=plt.gcf()
    p.gca().add_artist(my_circle)
    plt.axis('equal')
    plt.tight_layout()
    plt.xlabel(rating, labelpad=-175, fontsize=100, fontweight='bold')
    plt.savefig(f'/Users/blakeduncan/Downloads/Rating SVG/{rating}.svg', transparent=True)
    plt.show()
    
    rating = rating+1
```
